```python
time_data = [1, 2.5, 3, 4, 5.5, 6, 2, 4.5, 7]

low = []
moderate = []
high = []

for time in time_data:
    if time < 3:
        low.append(time)
    elif 3 <= time <= 5:
        moderate.append(time)
    else:
        high.append(time)

print("Low study hours (<3):", low)
print("Moderate study hours (3–5):", moderate)
print("High study hours (>5):", high)
```

```python
def sum_nested_list(nested_list):
    total = 0
    for item in nested_list:
        if isinstance(item, list):
            total += sum_nested_list(item)
        else:
            total += item
    return total
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]

result = sum_nested_list(nested_list)
print("Total sum of nested list:", result)
```

8) Task 2

```python
def generate_permutations(s):
    def backtrack(path, remaining, result):
        if not remaining:
            result.append(path)
            return
        seen = set()
        for i in range(len(remaining)):
            if remaining[i] in seen:
                continue
            seen.add(remaining[i])
            backtrack(path + remaining[i], remaining[:i] + remaining[i+1:], result)

    result = []
    backtrack("", s, result)
    return result

print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
file_system = {
    "root": {
        "file1.txt": 10,
        "file2.txt": 20,
        "subdir1": {
            "file3.txt": 30,
            "subsubdir1": {
                "file4.txt": 40
            }
        },
        "subdir2": {
            "file5.txt": 50
        }
    }
}
def calculate_directory_size(directory):
    total_size = 0
    for name, content in directory.items():
        if isinstance(content, dict):
            total_size += calculate_directory_size(content)
        else:
            total_size += content
    return total_size

root_size = calculate_directory_size(file_system["root"])
print(f"Total size of root directory: {root_size} KB")
```

```python
def min_coins(coins, amount):

    dp = [amount + 1] * (amount + 1)
    dp[0] = 0

    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != amount + 1 else -1


coins = [1, 2, 5]
amount = 11
print(min_coins(coins, amount))
```

```python
def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]


s1 = "abcde"
s2 = "ace"
print(longest_common_subsequence(s1, s2))
```

9) Task 3

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))
```
