```python
#task4
import numpy as np

A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

# 1
invA = np.linalg.inv(A)
print("\n1.A * A^-1 =\n",invA)

#2
AB = A @ B
BA = B @ A
print("\n2.\nAB =\n", AB)
print("BA =\n", BA)
print("Is AB == BA? ->", np.allclose(AB, BA))

#3
lhs = (A @ B).T
rhs = B.T @ A.T
print("\n3.\n(AB)^T =\n", lhs)
print("B^T * A^T =\n", rhs)
print("Is (AB)^T = B^T A^T ? ->", np.allclose(lhs, rhs))

#Solve Linear System using Inverse Method
M = np.array([[2, -3,  1],
              [1, -1,  2],
              [3,  1, -1]])

b = np.array([-1, -3, 9])

#4
invM = np.linalg.inv(M)
X1 = invM @ b
print("\n4.Solution using inverse:")
print("X =", X1)

#5
X2 = np.linalg.solve(M, b)
print("\n5.Solution using np.linalg.solve:")
print("X =", X2)
```

```python
#4.2
#1
import time

size = 1000000

print(f"Comparing Python lists vs NumPy for element-wise multiplication")
print(f"Size: {size:,}\n")

python_list_1 = list(range(size))
python_list_2 = list(range(size))

start = time.time()
python_mult = [a * b for a, b in zip(python_list_1, python_list_2)]
python_time = time.time() - start

print(f"Python lists: {python_time:.6f} seconds")

numpy_array_1 = np.arange(size)
numpy_array_2 = np.arange(size)

start = time.time()
numpy_mult = numpy_array_1 * numpy_array_2
numpy_time = time.time() - start

print(f"NumPy arrays: {numpy_time:.6f} seconds")
print(f"\n NumPy is {python_time / numpy_time:.2f}x FASTER!")
```

```python
#3
size_matrix = 1000

print(f"Comparing Python lists vs NumPy for matrix multiplication")
print(f"Matrix size: {size_matrix}×{size_matrix}\n")

python_matrix_1 = [[i+j for j in range(size_matrix)] for i in range(size_matrix)]
python_matrix_2 = [[i+j for j in range(size_matrix)] for i in range(size_matrix)]

print("⏳ Computing Python lists (this may take a while)...")
start = time.time()
python_matmul = [[sum(python_matrix_1[i][k] * python_matrix_2[k][j]
                      for k in range(size_matrix))
                  for j in range(size_matrix)]
                 for i in range(size_matrix)]
python_time = time.time() - start

print(f"Python lists: {python_time:.6f} seconds")

numpy_matrix_1 = np.random.rand(size_matrix, size_matrix)
numpy_matrix_2 = np.random.rand(size_matrix, size_matrix)

print("⏳ Computing NumPy arrays...")
start = time.time()
numpy_matmul = np.dot(numpy_matrix_1, numpy_matrix_2)
numpy_time = time.time() - start

print(f"NumPy arrays: {numpy_time:.6f} seconds")
print(f"\n NumPy is {python_time / numpy_time:.2f}x FASTER!")
print("\n This is why NumPy is so important for data science!")
```
